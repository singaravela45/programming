```python
import heapq
from collections import Counter
text = input("Enter string: ")
freq = Counter(text)
heap = [[weight, [char, ""]] for char, weight in freq.items()]
heapq.heapify(heap)
while len(heap) > 1:
    lo = heapq.heappop(heap)
    hi = heapq.heappop(heap)
    for pair in lo[1:]:
        pair[1] = '0' + pair[1]
    for pair in hi[1:]:
        pair[1] = '1' + pair[1]
    heapq.heappush(heap, [lo[0] + hi[0]] + lo[1:] + hi[1:])
huffman_codes = sorted(heapq.heappop(heap)[1:], key=lambda p: (len(p[-1]), p))
print("Huffman Codes:")
for char, code in huffman_codes:
    print(char, ":", code)
encoded = ""
code_dict = dict(huffman_codes)
for ch in text:
    encoded += code_dict[ch]
print("Encoded string:", encoded)
original_bits = len(text) * 8
compressed_bits = len(encoded)
compression_ratio = original_bits / compressed_bits
print("Original bits:", original_bits)
print("Compressed bits:", compressed_bits)
print("Compression Ratio:", compression_ratio)
```
