```python
#23BEC0460
queue = []
def enqueue(value):
    queue.append(value)
def dequeue():
    if len(queue) > 0:
        queue.pop(0)
def show_queue():
    print(queue)
def find_value(value):
    position = -1
    for i in range(len(queue)):
        if queue[i] == value:
            position = i
            break
    if position == -1:
        print("Value not found in queue")
    else:
        print("Value found at index:", position)
enqueue(7)
enqueue(14)
enqueue(21)
enqueue(28)
enqueue(35)
print("Queue before removing elements:")
show_queue()
dequeue()
dequeue()
print("Queue after removing elements:")
show_queue()
find_value(21)
```
