#### stack

```python
stack = []                          # create a stack using list
stack.append(x)                    # push element onto stack
stack.pop()                        # pop element from stack
class Stack:                       # user-defined stack
    def __init__(self):
        self.items = []            # initialize stack
    def push(self, value):
        self.items.append(value)   # push element
    def pop(self):
        if self.is_empty():
            raise Exception("Stack is empty")  # error if empty
        return self.items.pop()    # pop element
    def peek(self):
        if self.is_empty():
            raise Exception("Stack is empty")  # error if empty
        return self.items[-1]      # view top element
    def is_empty(self):
        return len(self.items) == 0  # check if stack is empty
    def size(self):
        return len(self.items)     # get stack size
```

#### linked List


```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

node1 = Node(10)

def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev=None
        curr=head
        while curr:
            temp=curr.next
            curr.next=prev
            prev=curr
            curr=temp
        return prev
```
